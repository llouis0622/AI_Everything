```python
X = []
for i in range(n_row - 1):
    row = []
    for j in range(n_col - 1):
        val = data[i][j]
        row.append(val)
    X.append(row)

y = []
for i in range(n_row - 1):
    val = data[i][10]
    y.append(val)
```

```python
n = len(X)
p = len(X[0])
```

```python
print(n)
print(p)
```

## 각 열의 평균 구하기

```python
mean_col = []
for j in range(p):
    sum_val = 0
    for i in range(n):
        sum_val += X[i][j]
    mean_val = (1 / n) * sum_val
    mean_col.append(mean_val)
```

```python
mean_col
```

## 공분산 행렬 구하기

### (X - mean) 행렬 구하기 

```python
mean_mat = []
for i in range(n):
    row = []
    for j in range(p):
        val = mean_col[j]
        row.append(val)
    mean_mat.append(row)
```

```python
mean_mat[0]
```

```python
mean_mat[1]
```

```python
X_minus_mean = la.subtract(X, mean_mat)
```

```python
X_minus_mean_t = la.transpose(X_minus_mean)
```

```python
res = la.matmul(X_minus_mean_t, X_minus_mean)
```

```python
res[0]
```

```python
cov_X = []
for i in range(p):
    row = []
    for j in range(p):
        val = (1 / n) * res[i][j]
        row.append(val)
    cov_X.append(row)
```

```python
# 트레이닝/테스트 데이터 분할
from sklearn.model_selection import train_test_split

X_tn, X_te, y_tn, y_te = train_test_split(X, y, random_state=0)
```

```python
# 데이터 표준화
from sklearn.preprocessing import StandardScaler

std_scale = StandardScaler()
std_scale.fit(X_tn)
X_tn_std = std_scale.transform(X_tn)
X_te_std = std_scale.transform(X_te)
```

```python
# KNN 회귀분석 학습
from sklearn.neighbors import KNeighborsRegressor

clf_knn = KNeighborsRegressor(n_neighbors=5)
clf_knn.fit(X_tn_std, y_tn)
```

```python
# 예측
pred_y = clf_knn.predict(X_te_std)
```

```python
pred_y
```

```python
# 모형 평가-MSE
from sklearn.metrics import mean_squared_error

print(mean_squared_error(y_te, pred_y))
```
